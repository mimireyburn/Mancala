```python
# Convert defaultdict to a more suitable data structure for plotting
def dict_to_df(q_table):
    states = list(q_table.keys())
    actions = list(range(len(q_table[states[0]])))
    df = pd.DataFrame(index=states, columns=actions)
    
    for state in states:
        for action in actions:
            df.at[state, action] = q_table[state][action]
    
    return df
```

```python
# Load the q_table.pkl file
with open('q_table.pkl', 'rb') as f:
    q_table = pickle.load(f)

df = dict_to_df(q_table)
```

```python
sns.heatmap(df.astype(float), cmap="YlGnBu")
```
